````python
import importlib

def print_package_version(package_name: str):
  try:
    package = importlib.import_module(package_name)

    if hasattr(package, "__version__"):
      print(f"{package_name} version: {package.__version__}")
    else:
      print(f"Package '{package_name}' does not have a __version__ attribute.")
  except ModuleNotFoundError:
    print(f"Package '{package_name}' is not installed.")
  except Exception as e:
    print(f"Error while checking version of '{package_name}': {e}")
````

To resolve the error, AI suggested to build a Wrapper class that intercepts the call. We tried it out but in the end, the same mistake arised.

````python
import torch

class RotaryWrapper(torch.nn.Module):
    def __init__(self, model):
        super().__init__()
        self.model = model

    def forward(self, *args, **kwargs):
        # Inject rotary embeddings before forward
        if "position_embeddings" not in kwargs or kwargs["position_embeddings"] is None:
            seq_len = kwargs["input_ids"].shape[1]
            kwargs["position_embeddings"] = self.model.model.rotary_emb(seq_len)
        return self.model(*args, **kwargs)
````
